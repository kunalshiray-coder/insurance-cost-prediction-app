```python

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("insurance.csv")
```

```python
# ============================================================
# Export Tableau-ready dataset
# ============================================================

tableau_df = df.copy()

# BMI
if "BMI" not in tableau_df.columns:
    tableau_df["BMI"] = tableau_df["Weight"] / ((tableau_df["Height"] / 100) ** 2)

# BMI Category
tableau_df["BMI_Category"] = pd.cut(
    tableau_df["BMI"],
    bins=[0, 18.5, 25, 30, 100],
    labels=["Underweight", "Normal", "Overweight", "Obese"],
    right=False
)

# Age Group
tableau_df["AgeGroup"] = pd.cut(
    tableau_df["Age"],
    bins=[17, 25, 35, 45, 55, 66],
    labels=["18-25", "26-35", "36-45", "46-55", "56-66"]
)

# Total Health Risks
tableau_df["TotalHealthRisks"] = tableau_df[
    [
        "Diabetes",
        "BloodPressureProblems",
        "AnyTransplants",
        "AnyChronicDiseases",
        "KnownAllergies",
        "HistoryOfCancerInFamily"
    ]
].sum(axis=1)

# Friendly Yes/No fields for Tableau visuals
binary_cols = [
    "Diabetes",
    "BloodPressureProblems",
    "AnyTransplants",
    "AnyChronicDiseases",
    "KnownAllergies",
    "HistoryOfCancerInFamily"
]

for col in binary_cols:
    tableau_df[col + "_Label"] = tableau_df[col].map({0: "No", 1: "Yes"})

# Save file
tableau_df.to_csv("insurance_tableau_ready.csv", index=False)

print("Tableau-ready file exported successfully: insurance_tableau_ready.csv")
```
